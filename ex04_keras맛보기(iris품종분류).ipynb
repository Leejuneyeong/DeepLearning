```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 학습용 iris 데이터 로드
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
# data : 실제 문제 데이터
# target : 정답 데이터
```

```python
X = iris.data # iris[data] 도 됨
y = iris.target 
```

#### 다중분류 문제에서는 정답데이터를 원핫인코딩 시켜줘야함
 - 모델이 예측한 확률 값이랑 실제 정답의 scale(범위)
 가 같아야 제대로 오차를 출력해낼 수 있고, 제대로 오차를 출력해야 신경망이 바르게 학습될 수 있기 때문!

```python
# 딥러닝 대표 라이브러리인 tensorflow에서 지원해주는 원핫인코딩 함수
from tensorflow.keras.utils import to_categorical
```

```python
# 배열(array) 형태로 원핫인코딩
y_one_hot = to_categorical(y)
y_one_hot
```

```python
# DataFrame 형태로 원핫인코딩 
y_one_hot = pd.get_dummies(y) 
y_one_hot
```

```python
# 학습 , 평가 데이터 분리
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, 
                                                    test_size=0.2,
                                                    random_state = 3
                                                    )
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### 딥러닝 신경망 모델링
- 입력 데이터의 특성 수
- 출력층 활성화 함수, 뉴런의 개수
- loss값 설정(categorical_crossentropy)
- optimizer : Adam,SGD

```python
# 설계 부터 학습 까지 스스로 진행
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
# 1. 신경망 구조 설계
model = Sequential() # 틀을 만듬

# 입력층(input_dim) + 중간층(Dense(units))
# input_dim : 특성(입력) 개수 만큼
# activation : 활성화 함수는 Dense 함수 안의 매개변수 값으로 지정
model.add(Dense(units = 200, input_dim = 4, activation='sigmoid')) # 들어올 수 있는 데이터 층
#Dense = 뉴런층(200개)

# 중간층
model.add(Dense(units = 100, activation = 'sigmoid')) # 층을 쌓음 add(Dense(units=100)) 100개
model.add(Dense(units = 50, activation = 'sigmoid')) #50개를 쌓음

# 출력층 -> y_train의 크기 개수만큼 해야댐
# 이진분류 !!! 유닛 = 1, activation = 시그모이드
# 출력층 3개 의미 : 클레스 레이블 개수만큼 해야댐
model.add(Dense(units = 3, activation='sigmoid'))

# 모델 정보
model.summary()
```

```python
# 2. 학습/평가방법 설정
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'SGD', # 최적화 함수 : 확률적 경사하강법
              metrics = ['accuracy']) #평가방법 설정(metrics) : 정확도
```

```python
h = model.fit(X_train, y_train, epochs = 1000) # epochs 모델 학습 횟수
```

```python
# 시각화 - 학습횟수에 따른 정확도 변화 그래프
plt.figure(figsize = (15,5))
plt.plot(
         h.history['accuracy'],
         label ='acc')
plt.legend()
plt.xlabel('epochs') # epochs = 학습횟수
plt.ylabel('acc')
plt.show()
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.evaluate(X_test, y_test) # 평가 , 97퍼센트 -> 모델을 잘 만듬
```
